# Introducción al uso del Jupyter Notebook

El **Jupyter Notebook** (*Cuaderno Jupyter*) es una herramienta que se puede utilizar para facilitar la resolución de (algunos o muchos) problemas de las guías de trabajos prácticos de la asignatura Radiodiagnóstico y Radioterapia (FI-UNER). **Su fortaleza radica en la posibilidad de combinar texto enriquecido (títulos, negrita, ecuaciones, etc) con código ejecutable**, intercalar imágenes y videos, etc. Desde el mismo notebook se pueden realizar presentaciones animadas (con diapositivas), sin tener que duplicar el contenido en otro archivo con diferente formato. Además el *notebook* se puede exportar como PDF, entre otros formatos.

Lo vamos a utilizar con Python, que es un lenguaje muy popular en la comunidad científica, de muy facil sintaxis y legibilidad, y para el cual existe un vasto universo de librerías que facilitan la resolución de ecuaciones con unidades físicas y la graficación de resultados.

Hay varias formas distintas de utilizar el Jupyter Notebook. Hay servicios online como Google Colab (https://colab.research.google.com) que permiten su uso remoto, pero nosotros vamos a usarlo ejecutándolo localmente. En ambos casos se accede a través de un navegador. ### ¿Qué es un archivo .ipynb?

## Celdas y Kernel
Hay dos términos bastante destacados que deberías notar y que probablemente sean nuevos para ti: las celdas y los 'kernel' son fundamentales tanto para comprender Jupyter como para entender por qué es algo más que un simple procesador de palabras. 

- Un Kernel es un "motor computacional" que ejecuta el código contenido en un documento de cuaderno.

- Una celda es un contenedor para texto que se muestra en el cuaderno o código que es ejecutado por el kernel del cuaderno.

### Ejecución de celdas

Cada celda se puede ejecutar con Shift+Enter (o con el botón Run). El navegador envía el contenido de la celda al servidor, quien procesa el código Python o Markdown y devuelve el resultado, que se imprime a continuación.

Al ejecutar una celda **Markdown** se procesan los códigos que identifican el texto enriquecido (negrita, listas, tablas, etc) y se muestra el texto con el formato adecuado. Si se ejecuta una celda **Code** se procesa el código Python y, de no mediar errores, se muestra el resultado.

## Trust notebook

https://jupyter-server.readthedocs.io/en/stable/operators/security.html#security-in-notebook-documents

La opción **"Trust Notebook"** en Jupyter Notebook permite marcar un cuaderno (notebook) como confiable después de ejecutar todo el código en él. Esto es útil en situaciones en las que ha descargado un cuaderno de una fuente externa y desea asegurarse de que **no haya código malicioso antes de ejecutarlo**. Cuando confía en un cuaderno, se supone que el código en él es seguro para ejecutar.

Sin embargo, el principal **inconveniente al usar la opción "Trust Notebook"** es:

- Seguridad: Marcar un cuaderno como confiable significa que confía en todo el código que contiene. Si el cuaderno proviene de una fuente no confiable o contiene código que no comprende completamente, podría ejecutar código malicioso sin darse cuenta. Por lo tanto, **debe ser cauteloso al confiar en cuadernos de fuentes no verificadas**.

Por otra parte, las **desventajas de los "untrusted notebooks"** en Jupyter pueden ser:

1. No se pueden **ejecutar automáticamente**: Cuando un cuaderno se marca como no confiable, Jupyter Notebook no permite ejecutar celdas de código automáticamente. Esto significa que, si tiene un cuaderno con muchas celdas de código, debe ejecutar cada celda manualmente, lo que puede ser tedioso y propenso a errores.

2. No se pueden **exportar a algunos formatos**: Algunos formatos de exportación, como HTML, PDF o slides, pueden estar restringidos para cuadernos no confiables. Esto puede limitar su capacidad para compartir o presentar el contenido del cuaderno de la manera deseada.

Por lo tanto, recuerda verificar el contenido manualmente. Antes de ejecutar un cuaderno no confiable, debes revisar cuidadosamente su contenido y el código que contiene.
